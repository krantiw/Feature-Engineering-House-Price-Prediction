# Advanced Housing Prices- Feature Engineering
The main aim of this project is to predict the house price based on various features

##### We will be performing the below steps in Feature Engineering

##### 1.Missing values
##### 2.Temporal variables
##### 3.Categorical variables: remove rare label

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option("display.max_columns",None) #to display the columns in the data fame
```

```python
dataset = pd.read_csv("train.csv")
dataset.head()
```

# 1.Missing values

#### Catergorical feature containing missing values

```python
categorical_feature__nan = [feature for feature in dataset.columns 
                       if dataset[feature].dtypes == "O" and dataset[feature].isnull().sum()>1]
categorical_feature__nan
```

```python
for feature in categorical_feature__nan:
    print(feature, ": Missing Values:{}%".format( np.round(dataset[feature].isnull().mean(), 4)))
```

#### Replace the nan values with a new category

```python
def replace_cat_feature(dataset, categorical_feature_nan):
    data = dataset.copy()
    data[categorical_feature_nan] = data[categorical_feature_nan].fillna("Missing")
    return data
```

```python
dataset = replace_cat_feature(dataset,categorical_feature__nan)
dataset[categorical_feature__nan].isnull().sum()
```

#### Numerical feature containing missing values

```python
numerical_feature_nan = [feature for feature in dataset.columns 
                         if dataset[feature].dtypes != "O" and dataset[feature].isnull().sum()>1 ]
numerical_feature_nan
```

```python
for feature in numerical_feature_nan:
    print(feature, ": Missing Value: {}%".format(np.round(dataset[feature].isnull().mean(),4)))
```

#### Replace the nan values with a median since there are outliers

```python
for feature in numerical_feature_nan:
    median_value = dataset[feature].median()
    
    dataset[feature + "nan"] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace= True)
dataset[numerical_feature_nan].isnull().sum()
```

```python
dataset
```

# 2.Temporal variables (Date-Time variables)

```python
#for how many years
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset["YrSold"]- dataset[feature]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

#### Numerical variables: Since the numerical variables are skewed we will perform log normal distribution

```python
import numpy as np
num_feature = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_feature:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

# 3.Categorical variables: remove rare labels

##### Remove categorical variables that are present less than 1% of the total observations
##### and convert that into a new label 

```python
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes == "O"]
categorical_feature
```

```python
for feature in categorical_feature:
    temp = dataset.groupby(feature)["SalePrice"].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], "Rare-Variable")
```
